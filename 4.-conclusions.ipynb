```python
import codecs
import pandas as pd
import numpy as np
```

```python
def uploudFile(file_name: str, file_path: str, destination_path: str) ->None:
  try:
    # Leemos el contenido del archivo como una cadena de texto
    with codecs.open(f'{file_path}{file_name}', 'r', encoding='utf-8') as f:
        file_content = f.read()
    # Escribe el contenido del archivo en un nuevo archivo con la misma ruta
    with codecs.open(destination_path, 'w', encoding='utf-8') as f:
        f.write(file_content)
    print('[INFO]Archivo cargado correctamente')
  except FileNotFoundError:
    print('[ERROR]No se ha podido realizar la carga del archivo')  
```

```python
# Define la ruta del archivo en tu Google Drive, la ruta de destino y el nombre del archivo
file_path = '/content/drive/MyDrive/NLP_practica/'  
file_name = 'graphics_nlp.py'
destination_path = '/content/graphics_nlp.py'
# Lee el contenido del archivo como una cadena de texto

uploudFile(file_name= file_name, file_path= file_path,destination_path= destination_path)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

def plot_model_metrics(metrics_df):
    fig, axs = plt.subplots(2, 2, figsize=(20, 20))

    # Gráfico de barras de accuracy
    sns.barplot(x='name_model', y='accuracy_train', data=metrics_df, color='blue', label='Train', ax=axs[0, 0])
    sns.barplot(x='name_model', y='accuracy_test', data=metrics_df, color='orange', label='Test', ax=axs[0, 0])
    axs[0, 0].set_xticklabels(axs[0, 0].get_xticklabels(), rotation=45, fontsize= 15)
    axs[0, 0].set_title('Accuracy de los modelos en conjunto de entrenamiento y prueba', fontsize= 20)
    axs[0, 0].legend()

    # Gráfico de barras de precision
    sns.barplot(x='name_model', y='precision_train', data=metrics_df, color='blue', label='Train', ax=axs[0, 1])
    sns.barplot(x='name_model', y='precision_test', data=metrics_df, color='orange', label='Test', ax=axs[0, 1])
    axs[0, 1].set_xticklabels(axs[0, 1].get_xticklabels(), rotation=45, fontsize= 15)
    axs[0, 1].set_title('Precision de los modelos en conjunto de entrenamiento y prueba', fontsize= 20)
    axs[0, 1].legend()

    # Gráfico de barras de recall
    sns.barplot(x='name_model', y='recall_train', data=metrics_df, color='blue', label='Train', ax=axs[1, 0])
    sns.barplot(x='name_model', y='recall_test', data=metrics_df, color='orange', label='Test', ax=axs[1, 0])
    axs[1, 0].set_xticklabels(axs[1, 0].get_xticklabels(), rotation=45, fontsize= 15)
    axs[1, 0].set_title('Recall de los modelos en conjunto de entrenamiento y prueba', fontsize= 20)
    axs[1, 0].legend()

    # Gráfico de barras de f1-score
    sns.barplot(x='name_model', y='f1_train', data=metrics_df, color='blue', label='Train', ax=axs[1, 1])
    sns.barplot(x='name_model', y='f1_test', data=metrics_df, color='orange', label='Test', ax=axs[1, 1])
    axs[1, 1].set_xticklabels(axs[1, 1].get_xticklabels(), rotation=45, fontsize= 15)
    axs[1, 1].set_title('F1-score de los modelos en conjunto de entrenamiento y prueba', fontsize= 20)
    axs[1, 1].legend()

    plt.tight_layout()
    plt.show()

```

```python
top_3 = metrics_models_reg_boos_df.nlargest(3, 'accuracy_test')
```

```python
plot_model_metrics(metrics_models_reg_boos_df)
```
